```python
import re
import numpy as np
import matplotlib.pyplot as plt

from scipy.spatial.distance import cosine
```

```python
with open('sentences.txt') as f:
    lines = [re.split('[^a-z]', line.lower()) for line in f]

lines = [[word for word in line if word != '']
            for line in lines]
lines
```

```python
word_set = {word for line in lines for word in line}
word_dict = {word: num for num, word in enumerate(word_set)}
word_dict
```

```python
matrix = []
for line in lines:
    text = [0] * len(word_dict)
    for word in line:
        text[word_dict[word]] += 1
    matrix.append(text)
matrix = np.array(matrix)
matrix
```

```python
cosines = [0] + [cosine(matrix[0, :], elem) for elem in matrix[1:, :]]
```

```python
with open('ans_1', 'w') as f:
    f.write('{} {}'.format(*np.argsort(cosines)[1:3]))
```

```python
x = np.array([1, 4, 10, 15])
x_3 = np.c_[np.ones(x.shape[0]),
            x,
            x**2,
            x**3]
x_3
```
