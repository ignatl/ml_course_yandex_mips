```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

from scipy.optimize import minimize, differential_evolution
```

# Задача 1. Минимизация гладкой функции

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
minimums = [minimize(f, x)['x'] for x in np.linspace(1, 30, 30)]
minimums
```

```python
ans_1_1 = round(minimize(f, 2, method='bfgs')['fun'], 2)
ans_1_2 = round(minimize(f, 30, method='bfgs')['fun'], 2)
```

```python
with open('ans_1', 'w') as out:
    out.write('{} {}'.format(ans_1_1, ans_1_2))
```

```python
plt.figure(figsize=(16, 10))
x_draw = np.linspace(1, 30, 291)
plt.plot(x_draw, f(x_draw))
```

# Задача 2. Глобальная оптимизация

```python
with open('ans_2', 'w') as out:
    out.write(str(round(differential_evolution(f, [(1, 30)])['fun'][0], 2)))
```

# Задача 3. Минимизация негладкой функции

```python
def h(x):
    return f(x).astype('int')
```

```python
plt.figure(figsize=(16, 10))
x_draw = np.linspace(1, 30, 291)
plt.plot(x_draw, h(x_draw))
```

```python
ans_3_1 = int(minimize(h, 30, method='bfgs')['fun'][0])
ans_3_2 = int(differential_evolution(h, [(1, 30)])['fun'])
ans_3_1, ans_3_2
```

```python
with open('ans_3', 'w') as out:
    out.write('{} {}'.format(ans_3_1, ans_3_2))
```
