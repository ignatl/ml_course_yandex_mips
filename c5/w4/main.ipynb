```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('coursera_sessions_train.txt', sep=';', header=None, names=['watched', 'bought'])
df_train.head()
```

```python
df_test = pd.read_csv('coursera_sessions_test.txt', sep=';', header=None, names=['watched', 'bought'])
df_test.head()
```

```python
from collections import Counter

watched_flat = [int(num) for elem in df_train['watched'] for num in elem.split(',')]
watched_freq = Counter(watched_flat)
watched_freq
```

```python
df_train = df_train[df_train['bought'].notnull()]
df_train.shape
```

```python
df_test = df_test[df_test['bought'].notnull()]
df_test.shape
```

```python
bought_flat = [int(num) for elem in df_train['bought'] for num in elem.split(',')]
bought_freq = Counter(bought_flat)
bought_freq
```

```python
watched_rate = sorted(watched_freq, key=watched_freq.get, reverse=True)
bought_rate = sorted(bought_freq, key=bought_freq.get, reverse=True)
```

```python
def sort_list_by_watched(l):
    return sorted(set(l), key=lambda x: watched_freq.get(int(x), 0), reverse=True)

def sort_list_by_bought(l):
    return sorted(set(l), key=lambda x: bought_freq.get(int(x), 0), reverse=True)

def average_precision(k, df, mode):
    precisions = []
    for index, row in df.iterrows():
        if mode == 'watched':
            recomendations = set(sort_list_by_watched(row['watched'])[:k])
        elif mode == 'bought':
            recomendations = set(sort_list_by_bought(row['watched'])[:k])
        precision = len(recomendations.intersection(row['bought'])) / min(k, len(recomendations))
        precisions.append(precision)
    return round(sum(precisions) / len(precisions), 2)

def average_recall(k, df, mode):
    precisions = []
    for index, row in df.iterrows():
        if mode == 'watched':
            recomendations = set(sort_list_by_watched(row['watched'])[:k])
        elif mode == 'bought':
            recomendations = set(sort_list_by_bought(row['watched'])[:k])
        precision = len(recomendations.intersection(row['bought'])) / len(row['bought'])
        precisions.append(precision)
    return round(sum(precisions) / len(precisions), 2)
```

```python
average_precision(5, df_train, 'watched')
print(sort_list_by_watched(df_train['watched'][7]))
print(df_train['bought'][7])
```

```python
with open('ans_1', 'w') as fout:
    fout.write(' '.join([str(average_recall(1, df_train, 'watched')),
                         str(average_precision(1, df_train, 'watched')),
                         str(average_recall(5, df_train, 'watched')),
                         str(average_precision(5, df_train, 'watched'))]))

with open('ans_2', 'w') as fout:
    fout.write(' '.join([str(average_recall(1, df_test, 'watched')),
                         str(average_precision(1, df_test, 'watched')),
                         str(average_recall(5, df_test, 'watched')),
                         str(average_precision(5, df_test, 'watched'))]))

with open('ans_3', 'w') as fout:
    fout.write(' '.join([str(average_recall(1, df_train, 'bought')),
                         str(average_precision(1, df_train, 'bought')),
                         str(average_recall(5, df_train, 'bought')),
                         str(average_precision(5, df_train, 'bought'))]))

with open('ans_4', 'w') as fout:
    fout.write(' '.join([str(average_recall(1, df_test, 'bought')),
                         str(average_precision(1, df_test, 'bought')),
                         str(average_recall(5, df_test, 'bought')),
                         str(average_precision(5, df_test, 'bought'))]))
```
