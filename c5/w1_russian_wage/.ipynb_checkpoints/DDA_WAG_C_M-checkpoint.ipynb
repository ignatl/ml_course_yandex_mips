# Прогнозирование уровня средней заработной платы в России

```python
%pylab inline
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings
from itertools import product
```

```python
data = pd.read_csv('WAG_C_M.csv',';', index_col=['month'], parse_dates=['month'], dayfirst=True)
```

```python
data.columns.values[0] = 'wage' # Средняя номинальная заработная плата (WAG_C_M) 
data.head() 
```

```python
plt.figure(figsize(15,7))
data.wage.plot()
plt.ylabel('wage')
pylab.show()
```

### 1. Визуальный анализ ряда. Проверка стационарности и STL-декомпозиция ряда:

```python
plt.figure(figsize(15,10))
sm.tsa.seasonal_decompose(data.wage).plot()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(data.wage)[1])
```


Четко прослеживается тренд: плавное долгосрочное увеличение уровня.
Четко прослеживается сезонность: циклическое изменение уровня с постоянным периодом.
Дисперсия значений в начале значительно меньше дисперсии значений в конце.

Что подтверждает критерий Дики-Фуллера, не отвергающий гипотезу о не стационарности.

### 2. Стабилизация дисперсии
Сделаем преобразование Бокса-Кокса для стабилизации дисперсии:

```python
data['boxcox'], lmbda = stats.boxcox(data.wage)
plt.figure(figsize(15,7))
data.boxcox.plot()
plt.ylabel(u'Transformed wages')
print("Оптимальный параметр преобразования Бокса-Кокса: %f" % lmbda)
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(data.boxcox)[1])
```

```python
data.boxcox.head()
```

Дисперсия стала более стабильной. Но критерий Дики-Фуллера всё ещё не отвергает гипотезу о нестационарности.

### 3. Дифференцирование в поисках стационарности

Попробуем сезонное дифференцирование. Попробуем сезонное дифференцирование по году (12 месяцев); сделаем на продифференцированном ряде STL-декомпозицию и проверим стационарность:

```python
data['wage_box_diff'] = data.boxcox - data.boxcox.shift(12)
plt.figure(figsize(15,10))
sm.tsa.seasonal_decompose(data.wage_box_diff[12:]).plot()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(data.wage_box_diff[12:])[1])
```

Критерий Дики-Фуллера отвергает гипотезу нестационарности, и визуально ряд выглядит лучше — тренда больше нет. 
Проведём еще одно дифференцирование:

```python
data['wage_box_diff_12_1'] = data.wage_box_diff - data.wage_box_diff.shift(1)
plt.figure(figsize(15,10))
sm.tsa.seasonal_decompose(data.wage_box_diff_12_1[13:]).plot() 
plt.show()
```

```python
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(data.wage_box_diff_12_1[13:])[1])
```

### 4. Выбор начальных приближений для p,q,P,Q.

Посмотрим на ACF и PACF полученного ряда:

```python
plt.figure(figsize(15,8))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(data.wage_box_diff_12_1[13:].values.squeeze(), lags=48, ax=ax)
pylab.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(data.wage_box_diff_12_1[13:].values.squeeze(), lags=48, ax=ax)
pylab.show()
```

Начальные приближения: Q=0, q=1, P=3, p=1.

```python
ps = range(0, 4)
d=1 
qs = range(0, 4)
Ps = range(0, 2)
D=1
Qs = range(0, 2)
```

```python
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

### 5. Обучение и сравнение моделей-кандидатов, выбор победителя.

```python
%%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for param in parameters_list:
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model=sm.tsa.statespace.SARIMAX(data.boxcox, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 12)).fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

```python
print(best_model.summary())
```

### 6. Анализ остатков построенной модели.

```python
plt.figure(figsize(15,8))
plt.subplot(211)
best_model.resid[13:].plot()
plt.ylabel(u'Residuals')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[13:].values.squeeze(), lags=48, ax=ax)

print("Критерий Стьюдента: p=%f" % stats.ttest_1samp(best_model.resid[13:], 0)[1])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(best_model.resid[13:])[1])
```

Остатки несмещены (подтверждается критерием Стьюдента) стационарны (подтверждается критерием Дики-Фуллера и визуально), неавтокоррелированы (подтверждается критерием Льюнга-Бокса и коррелограммой). Посмотрим, насколько хорошо модель описывает данные:

```python
def invboxcox(y,lmbda):
   if lmbda == 0:
      return(np.exp(y))
   else:
      return(np.exp(np.log(lmbda*y+1)/lmbda))
```

```python
data['model'] = invboxcox(best_model.fittedvalues, lmbda)
plt.figure(figsize(15,7))
data.wage.plot()
data.model[13:].plot(color='r')
plt.ylabel(u'Средняя номинальная заработная плата')
plt.xlabel(u'Период наблюдения')
plt.legend([u'Фактические данные', u'Смоделированые данные'], loc=1, ncol = 2, prop={'size':10})
pylab.show()
```
